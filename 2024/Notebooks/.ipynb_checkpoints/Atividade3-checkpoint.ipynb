```python
# Caminho do arquivo CSV
path = r'C:\Users\rafael.shibayama\Documents\GitHub\ENS410064\2024\Dados\chuvas_536012.csv'

# Carregar o arquivo CSV, especificando a codificação e delimitador corretos e ignorando linhas iniciais desnecessárias
data = pd.read_csv(path, encoding='ISO-8859-1', skiprows=15, delimiter=';')

# Renomear as colunas para facilitar o trabalho
data.columns = ['Station', 'Day', 'Month', 'Year', 'Precipitation']

# Criar uma coluna de data combinando as informações de dia, mês e ano
data['Date'] = pd.to_datetime(data[['Year', 'Month', 'Day']])

# Ordenar os dados pela data
data = data.sort_values('Date')

monthly_summary = data.resample('M', on='Date')['Precipitation'].agg(['min', 'max', 'mean']).reset_index()

# Supondo que 'monthly_summary' contenha a média de precipitação mensal e a coluna 'Date'
# Converter as datas em números para aplicar o estimador
monthly_summary['Date_ordinal'] = monthly_summary['Date'].map(lambda date: date.toordinal())

# Aplicar o estimador de Thiel-Sen
slope, intercept, lower_slope, upper_slope = theilslopes(monthly_summary['mean'], monthly_summary['Date_ordinal'], 0.95)

# Exibir os resultados
print(f"Inclinação (slope): {slope}")
print(f"Intercepto: {intercept}")
print(f"Intervalo de Confiança 95%: [{lower_slope}, {upper_slope}]")

# Plotar a série temporal com a linha de tendência
plt.figure(figsize=(10, 6))
plt.plot(monthly_summary['Date'], monthly_summary['mean'], label='Precipitação Média', color='blue')
plt.plot(monthly_summary['Date'], intercept + slope * monthly_summary['Date_ordinal'], 'r--', label='Tendência (Thiel-Sen)')
plt.xlabel('Data')
plt.ylabel('Precipitação (mm)')
plt.title('Tendência de Precipitação com Estimador de Thiel-Sen')
plt.legend()
plt.grid(True)

plt.figure(figsize=(12, 6))
sns.boxplot(x=data['Year-Month'].astype(str), y=data['Precipitation'])
plt.xticks(rotation=90)  # Rotacionar os rótulos para facilitar a leitura
plt.xlabel('Ano-Mês')
plt.ylabel('Precipitação (mm)')
plt.title('Distribuição Mensal de Precipitação')
plt.grid(True)

# Exibir o gráfico
plt.show()
```
